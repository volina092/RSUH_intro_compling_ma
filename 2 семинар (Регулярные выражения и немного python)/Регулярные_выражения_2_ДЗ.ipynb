## Lookahead/Lookbehind группы

Это группы "заглядывающие" вперёд или назад внутри текста без захвата фрагмента в match.

Заглядывающие вперёд (lookahead) задаются так: `r'X(?=Y)'`. Найдёт только X, за которым следует Y.

Заглядывающие назад (lookbehind) задаются так: `r'(?<=Y)X'`. Найдёт только X, который следует за Y.

```python
# найдёт только тот "тот", после которого идёт "самый"
re.findall(r'тот(?=\s+самый)', 'Ах, тот день! В тот самый день...')
```

```python
# заменит только тот "тот", после которого идёт "самый"
re.sub(r'тот(?=\s+самый)', 'ТОТ', 'Ах, тот день! В тот самый день...')
```

Существуют так же "негативные" lookahead/lookbehind группы. Они помогают найти только вхождения, без контекста "заглядывающей группы". То есть:

РЕ с **негативной заглядывающей вперёд (lookahead)** группой — `r'X(?!Y)'` — найдёт только X, за которым *не* следует Y.

РЕ с **негативной заглядывающей назад (lookbehind)** группой — `r'(?<!Y)X'` — найдёт только X, который *не* следует за Y.

Множественное число в грузинском языке образуется присоединением суффикса *-ებ-* к корню. Падежный суффикс (*-ი* для именительного падежа) "присоединяется" в самый конец словоформы — к показателю множественного числа или корню (если слово в единственном числе).

*სახლი - სახლ**ებ**ი (дом - дома)*

*ხიდი - ხიდ**ებ**ი (мост - мосты)*

*თევზი - თევზ**ებ**ი (рыба - рыбы)*



Если основа оканчивается на гласный (*უ*, *ო*, *ე*), в единственном числе показатель именительного падежа (*-ი*) отсутствует, но при образовании множественного числа он "появляется" (поскольку фонологическое правило применяется уже не к голой основе):

*საღამო - საღამო**ებ**ი (вечер - вечера)*

*დღე - დღე**ებ**ი (день - дни)*


При этом, если конечный гласный основы — *ა*, он "выпадает" при образовании множественного числа:

*დანა - დან**ებ**ი (нож - ножи)*
